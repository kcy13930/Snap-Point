```python
import get_info as gi
import librosa
import numpy as np
import matplotlib.pyplot as plt
```

```python
filepath = '30_times.wav'
gi.basic_info(filepath)
```

```python
y, sr = librosa.load(filepath)
plt.figure(figsize=(16, 5))
plt.xticks(rotation=30)
librosa.display.waveplot(y, sr=sr)
plt.show()
```

```python
i = 0 
cnt = 0
beat_frames = []
while i < len(y):
    if (y[i] > 0.25):
        print(i, y[i])
        cnt += 1
        beat_frames.append(i)
        i += 800
    i += 2
print(cnt)

beat_times = librosa.core.frames_to_time(beat_frames, sr=sr, hop_length=1)
print(beat_times)
```
