```python
import get_info as gi
import librosa
import matplotlib.pyplot as plt
```

```python
filepath = '30_times.wav'
gi.basic_info(filepath)
gi.specshow_graph(filepath)
```

```python
y, sr = librosa.load('30_times.wav', sr=20000)
y_harm, y_perc = librosa.effects.hpss(y)
plt.figure(figsize=(14, 5))
librosa.display.waveplot(y_harm, sr=sr, alpha=0.25)
librosa.display.waveplot(y_perc, sr=sr, color='r', alpha=0.5)
plt.title('Harmonic + Percussive')
plt.tight_layout()
plt.show()
```

```python
y, sr = librosa.load('30_times.wav')
y_mfcc = librosa.feature.mfcc(y=y, sr=sr)
print(y_mfcc.shape)
print(y.shape)
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(y_mfcc, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```
